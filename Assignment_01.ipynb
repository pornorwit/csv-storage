### 4) เขียน ส่วนของโปรแกรม python เพื่อตอบคำถามต่อไปนี้
ไฟล์: sales_dirty.csv มีปัญหา:
- quantity บางแถวว่าง
- price เป็น string เช่น "150"

```python
import pandas as pd

sales_dirty_path = "https://raw.githubusercontent.com/pornorwit/csv-storage/refs/heads/main/sales_dirty.csv"

# 4.1) อ่านด้วย pd.read_csv()
df_sales_dirty = pd.read_csv(sales_dirty_path, encoding="utf-8-sig")
print('[4.1]')
print('---- อ่านไฟล์ csv สำเร็จ ----')
```

```python
# 4.2) แทน quantity ที่ว่างด้วย 0 และแปลงเป็น int
print('[4.2]')
print("---- Before fillna & Conversion ----\n")
print(df_sales_dirty)
df_sales_dirty.info()

df_sales_dirty_FillNA = df_sales_dirty.fillna({"quantity" : 0})
df_sales_dirty_FillNA["quantity"] = df_sales_dirty_FillNA["quantity"].astype(int)

print("\n---- After fillna & Conversion ----\n")
print(df_sales_dirty_FillNA)
df_sales_dirty_FillNA.info()
```

```python
# 4.3) แปลง price เป็น float
print('[4.3]')
print("---- Before conversion ----")
df_sales_dirty_FillNA.info()

df_sales_dirty_FillNA["price"] = df_sales_dirty_FillNA["price"].astype(float)

print("\n---- After conversion ----")
df_sales_dirty_FillNA.info()
```

```python
# 4.4) Filter เฉพาะแถวที่ quantity > 0
df_sales_dirty_Filter = df_sales_dirty_FillNA[df_sales_dirty_FillNA['quantity'] > 0]
print('[4.4]')
print("---- quantity > 0 ----")
print(df_sales_dirty_Filter)
```
